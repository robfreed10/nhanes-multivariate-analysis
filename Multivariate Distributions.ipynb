```python
import numpy as np ###for getting distribution
import pandas as pd
import seaborn as sns; sns.set() ###for a different plotting theme
```

```python
import matplotlib.pyplot as plt ### for plotting
```

```python
r = -1
```

```python
###1 equal perfect upward sloping line
### -1 equal perfect downward sloping line
### 0 gets us a 'cloud' of points
```

```python
mean = [15, 5]
cov = [1, r], [r, 1]
x, y = x, y = np.random.multivariate_normal(mean, cov, 400).T
```

```python
plt.figure(figsize=(10,5))
```

```python
###Plot histograms of x and y next to each other 
plt.subplot(1,2,1)
plt.hist(x = x, bins = 15)
plt.title("X") ###Marginal Distribution of X

plt.subplot(1,2,2)
plt.hist(x = y, bins = 15)
plt.title("Y")###Marginal Distribution of Y

plt.show()
```

```python
###Plot the data
plt.figure(figsize=(10,10))
plt.subplot(2,2,2)
plt.scatter(x = x, y = y)
plt.title("Joint Distribution of X and Y")
plt.show()
```

```python
###Plotting the Marginal Distribution of X
plt.subplot(2,2,4)
plt.hist(x = x, bins = 15)
plt.title("Marginal Distribution of X")

###Plotting the Marginal Distribution of Y
plt.subplot(2,2,1)
plt.hist(x = y, orientation = "horizontal", color = "red", bins = 15)
plt.title("Marginal Distributions of Y")

plt.show()
```
